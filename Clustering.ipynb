DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("crime_data.csv")
df.head(10)
```

```python
df.info()
```

```python
newdf=df.iloc[:,1:5]
newdf.head(5)
```

```python
array = newdf.values
array
```

Normalising data

```python
stscaler = StandardScaler().fit(array)
x=stscaler.transform(array)
x
```

```python
db= DBSCAN(eps=2,min_samples=5)
db.fit(x)
db.labels_
```

```python
a=pd.DataFrame(db.labels_,columns=['cluster'])
pd.concat([df,a],axis=1)
```

Hierarchical

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
cm=pd.read_csv("crime_data.csv")
cm.head(5)
```

```python
cm.info()
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
cm_norm=norm_func(cm.iloc[:,1:])
cm_norm.head(5)
```

```python
den=sch.dendrogram(sch.linkage(cm_norm,method="average"))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="average")
hc
```

```python
y=hc.fit_predict(cm_norm)
y
```

```python
cm['cluster_id']=hc.labels_
cm.head(10)
```

Kmeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
cm_km=pd.read_csv("crime_data.csv")
```

```python
cm_km_norm=norm_func(cm_km.iloc[:,1:])
cm_km_norm.head(5)
```

```python
#elbow curve
l=[]
for i in range(1,11):
  km=KMeans(n_clusters=i)
  km.fit(cm_km_norm)
  l.append(km.inertia_)

plt.plot(range(1,11),l)
plt.xlabel("clusters")
plt.ylabel("l")
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(cm_km_norm)
y=model.labels_
y
```

```python
cm_km['cluster_id']=y
cm_km.head(10)
```
