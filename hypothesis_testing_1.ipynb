 Q1.
 A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.


```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import norm
```

```python
ht=pd.read_csv('/content/Cutlets.csv')
ht.head(10)
```

we perform 2 tail t test as the hypothesis case is
- h0 if (unit a = unit b)
- h1 if (unit a <> unit b)

```python
p_value=stats.ttest_ind(ht["Unit A"],ht["Unit B"])
print(p_value)
```

alpha = 0.05
 p= 0.47

 p < alpha
 0.47 < 0.05   [false]

 So it is null hypothesis

Q3. Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.

```python
ht3=pd.read_csv("/content/BuyerRatio.csv")
ht3.head()

```

```python
ht3_val=ht3.iloc[:,1:6]
ht3_val
```

Using Chi2 contengency technique

O/p is (t value, p value , data frame, observations)

```python
from scipy.stats.contingency import chi2_contingency
val=chi2_contingency(ht3_val)
p_val = val[1]
print(p_val)
```

- p < alpha

- 0.6 < 0.05  (not satisfied)

- Null hypothesis
